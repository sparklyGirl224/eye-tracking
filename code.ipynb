```python
%pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV file
df = pd.read_csv('[uxf_rig]_variant_Eye_tracking_data_2D_T001(in).csv')
df = df.dropna()

# Invert y-axis to match image coordinates
df['y2'] = df['coord_y'] * -1

# Calculate image extent based on coordinates
x_min, x_max = df['coord_x'].min(), df['coord_x'].max()
y_min, y_max = df['y2'].min(), df['y2'].max()

plt.figure(figsize=(10, 8))

# Create heatmap overlay
sns.kdeplot(
    x=df['coord_x'],
    y=df['y2'],
    cmap='coolwarm',
    fill=True,
    thresh=0.05,
    alpha=0.6,
    levels=20,
    antialiased=True
)

# Configure plot settings
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Eye Tracking Heatmap Overlay')
plt.grid(False)  # Typically better for image overlays


plt.show()
```
